### Initialize variables

```python
# dimensions of our images.
img_width, img_height = 150, 150
```

```python
if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)
```

### Import images

```python
def create_directory(dir):
    path = '/tmp/chest_xray'+ dir
    isExist = os.path.exists(path)
    if not isExist:
      os.makedirs(path)

train_data_dir = source_train_data_dir
validation_data_dir = source_validation_data_dir
test_data_dir = source_test_data_dir

create_directory(source_train_data_dir)
create_directory(source_validation_data_dir)
create_directory(source_test_data_dir)

paginator = s3_source.get_paginator('list_objects_v2')
train_objects_pages = paginator.paginate(Bucket=source_bucket, Prefix=train_data_dir)

nb_train_samples = 0

for train_objects_page in train_objects_pages:
    nb_train_samples += sum(1 for _ in train_objects_page['Contents'])
    for file in train_objects_page.get('Contents', []):
        filename = file.get('Key')
        dest_pathname = '/tmp/chest_xray/' + filename
        if not os.path.exists(os.path.dirname(dest_pathname)):
            os.makedirs(os.path.dirname(dest_pathname))
        s3_source.download_file(source_bucket, file.get('Key'), dest_pathname)

print(nb_train_samples)

validation_objects_pages = paginator.paginate(Bucket=source_bucket, Prefix=validation_data_dir)

nb_validation_samples = 0

for validation_objects_page in validation_objects_pages:
    nb_validation_samples += sum(1 for _ in validation_objects_page['Contents'])

print(nb_validation_samples)
```

### Create Sequential model

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
def s3_image_generator(files, batch_size = 16, preprocessor):
    s3 = boto3.resource('s3')
    while True:
        batch_paths = np.array(files)
        batch_input = []
        batch_output = [0] * len(files)
        for input_path in batch_paths:
            input = fetch_input(input_path, s3)
            input = preprocessor(input)
            batch_input += [ input ]
        batch_x = np.array( batch_input )
        batch_y = np.array( batch_output )
        yield( batch_x, batch_y )
```

### Upload images from the different sets

```python
# this is the augmentation configuration we will use for training
train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)
```

```python
# this is the augmentation configuration we will use for testing:
# only rescaling
test_datagen = ImageDataGenerator(rescale=1. / 255)
```

```python
train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')
```

```python
validation_generator = test_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')
```

```python
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')
```

### Fit the model

```python
model.fit_generator(
    train_generator,
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=nb_validation_samples // batch_size)
```

### Save the model (weights + complete model)

```python
model.save(destination_model_name)
```

### Evaluate the model

```python
# evaluate the model
scores = model.evaluate_generator(test_generator)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
